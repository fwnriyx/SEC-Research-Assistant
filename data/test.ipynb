```python
import pandas as pd

def display_revenue(ticker):
    stock = yf.Ticker(ticker)
    financials = stock.financials
    # print(financials.index)
    # print(financials.loc["Total Revenue"])

    if "Total Revenue" not in financials.index:
        return None

    revenue = stock.financials.loc["Total Revenue"]

    
    df = revenue.copy()
    df.index = pd.to_datetime(df.index)
    df = df.sort_index()
    print("hey")
    print(df)

    df = pd.DataFrame({
        "Date": pd.to_datetime(revenue.index),
        "Revenue (B)": revenue.values / 1e9
    })
    return df
ticker = yf.Ticker("NVDA")

print(display_revenue("NVDA"))
```
